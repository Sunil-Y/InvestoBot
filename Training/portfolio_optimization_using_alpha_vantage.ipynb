```python
from collections import OrderedDict
startDate = '2019-01-01'
endDate = '2019-11-22'
companyAllocations = OrderedDict([('MSFT',0.4), ('AAPL', 0.3), ('FB', 0.3)])
currentValue = 100000

googlTemp = googl.rename(columns={'Close':'GOOGL'})
msftTemp = msft.rename(columns={'Close':'MSFT'})
fbTemp = fb.rename(columns={'Close':'FB'})
googlTemp.drop(['Open', 'High', 'Low', 'Volume'], axis=1, inplace=True)
msftTemp.drop(['Open', 'High', 'Low', 'Volume'], axis=1, inplace=True)
fbTemp.drop(['Open', 'High', 'Low', 'Volume'], axis=1, inplace=True)
```

```python
norm = newdf/newdf.iloc[0]
norm = norm.astype(float)

norm['GOOGL'] = norm['GOOGL'].astype(float)
norm['MSFT'] = norm['MSFT'].astype(float)
norm['FB'] = norm['FB'].astype(float)

#norm['GOOGL'] *= float(companyAllocations.get('GOOGL')*currentValue)
norm['MSFT'] *= float(companyAllocations.get('MSFT')*currentValue)
norm['FB'] *= float(companyAllocations.get('FB')*currentValue)
evaluation = norm.sum(axis=1)
evaluation
newdf
```

```python
def printStats(evaluation):
    cumulativeReturn = (evaluation[-1]/evaluation[0])-1

    dailyReturn = evaluation.copy()
    dailyReturn[1:] = (evaluation[1:]/evaluation[:-1].values)-1
    dailyReturn = dailyReturn[1:]

    dailyReturnAverage = dailyReturn.mean()

    dailyReturnStd = dailyReturn.std()

    sharpeRatio = dailyReturnAverage/dailyReturnStd
    sharpeRatio = np.sqrt(len(evaluation))*sharpeRatio

    portValue = evaluation.iloc[len(evaluation)-1]
    
    print("Cumulative Return is ", cumulativeReturn)
    print("Daily Return mean is ", dailyReturnAverage)
    print("Volatility or Standard Deviation is ", dailyReturnStd)
    print("Sharpe Ratio is ", sharpeRatio)
    print("Portfolio evaluation at the end of the term is ", portValue)
    
    return cumulativeReturn, dailyReturn, dailyReturnAverage, dailyReturnStd, sharpeRatio, portValue
cumulativeReturn, dailyReturn, dailyReturnAverage, dailyReturnStd, sharpeRatio, portValue = printStats(evaluation)
```

```python
from scipy.optimize import minimize
def calculateSharpeRatio(allocs, prizes):
    norm = afterdf/afterdf.iloc[0]
    allocated = norm * allocs
    newValue = allocated * currentValue
    evaluationNew = newValue.sum(axis=1)
    
    dailyReturn = evaluationNew.copy()
    dailyReturn[1:] = (evaluationNew[1:]/evaluationNew[:-1].values)-1
    dailyReturn = dailyReturn[1:]
    sharpeRatio = dailyReturn.mean()/dailyReturn.std()
    sharpeRatio = np.sqrt(len(evaluationNew))*sharpeRatio
    return -1*sharpeRatio
```

```python
baseAllocation = np.array([0.33333,0.33333,0.33334])
alloc = minimize(calculateSharpeRatio, baseAllocation, args=(afterdf,),method='SLSQP',bounds=((0,1),(0,1),(0,1)),constraints=({'type':'eq','fun':lambda inputs: 1.0-np.sum(inputs)}))
```
